### 7. Display age statistics for each race (race feature) and each gender (sex feature). Use groupby() and describe(). Find the maximum age of men of Amer-Indian-Eskimo race.

Age statistics for each race and gender, maximum age of men of Amer-Indian-Eskimo race:

```python
for (race, sex), sub_df in data.groupby(["race", "sex"]):
    print("Race: {0}, sex: {1}".format(race, sex))
    print(sub_df["age"].describe())
```

### 8. Among whom is the proportion of those who earn a lot (>50K) greater: married or single men (marital-status feature)? Consider as married those who have a marital-status starting with Married (Married-civ-spouse, Married-spouse-absent or Married-AF-spouse), the rest are considered bachelors.

Checking data in <b>martial status:

```python
data.loc[data["salary"] == '>50K', 'marital-status'].value_counts()
```

Calculation proportion:

```python
proportion = data.loc[data["salary"] == '>50K', 'marital-status'].value_counts()
```

```python
(proportion['Widowed']+proportion['Never-married']+proportion['Divorced']+proportion['Separated'])/(proportion['Married-civ-spouse']+proportion['Married-spouse-absent']+proportion['Married-AF-spouse'])*100
```

### 9. What is the maximum number of hours a person works per week (hours-per-week feature)? How many people work such a number of hours, and what is the percentage of those who earn a lot (>50K) among them?

Maximum number of hours a person works per week:

```python
data['hours-per-week'].max()
```

Amount of people that work maximum number of hours:

```python
len(data[data['hours-per-week'] == 99])
```

Amount of people that work maximum number of hours and earn a lot (>50K):

```python
len(data[(data['hours-per-week'] == 99) & (data['salary']== '>50K')])
```

Percentage of those who earn a lot (>50K) among people that work 99 hours per week:

```python
len(data[(data['hours-per-week'] == 99) & (data['salary']== '>50K')])/len(data[data['hours-per-week'] == 99])*100
```

### 10. Count the average time of work (hours-per-week) for those who earn a little and a lot (salary) for each country (native-country). What will these be for Japan?

Filtering people by income:

```python
big_income = data.loc[data['salary'] == '>50K']
low_income = data.loc[data['salary'] == '<=50K']
```

Average time of work (hours-per-week) for those who earn a lot for each country:

```python
big_income.groupby('native-country').mean()['hours-per-week']
```

Average time of work (hours-per-week) for those who earn a little for each country:

```python
low_income.groupby('native-country').mean()['hours-per-week']
```

Average time of work (hours-per-week) for those who earn a lot in Japan:

```python
big_income.loc[big_income['native-country'] == 'Japan'].mean()['hours-per-week']
```

Average time of work (hours-per-week) for those who earn a little in Japan:

```python
low_income.loc[low_income['native-country'] == 'Japan'].mean()['hours-per-week']
```
